```python
import time, requests

ship = "TRAILJACK-1"
target = "COPPER_ORE"

def get_cargo():
    return requests.get(f"{BASE_URL}/my/ships/{ship}/cargo", headers=headers).json()

def jettison_non_copper():
    cargo = get_cargo()
    for item in cargo["data"]["inventory"]:
        if item["symbol"] != target:
            symbol, units = item["symbol"], item["units"]
            res = requests.post(
                f"{BASE_URL}/my/ships/{ship}/jettison",
                headers=headers,
                json={"symbol": symbol, "units": units}
            ).json()
            print(f"üóëÔ∏è  Jettisoned {units}x {symbol}")

while True:
    # Extract ore
    mine = requests.post(f"{BASE_URL}/my/ships/{ship}/extract", headers=headers).json()
    yield_info = mine.get("data", {}).get("extraction", {}).get("yield", {})
    mined_symbol = yield_info.get("symbol")
    mined_units = yield_info.get("units", 0)
    print(f"‚õèÔ∏è  Mined {mined_units}x {mined_symbol}")

    # Immediately drop anything that's not Copper
    if mined_symbol != target and mined_symbol:
        res = requests.post(
            f"{BASE_URL}/my/ships/{ship}/jettison",
            headers=headers,
            json={"symbol": mined_symbol, "units": mined_units}
        ).json()
        print(f"üóëÔ∏è  Dropped {mined_units}x {mined_symbol} immediately")

    # Check how much Copper we have
    cargo = get_cargo()
    inventory = {i["symbol"]: i["units"] for i in cargo["data"]["inventory"]}
    copper_units = inventory.get(target, 0)
    print(f"üö¢  Current Copper: {copper_units} / {cargo['data']['capacity']}")

    # Stop when cargo full or enough Copper collected
    if copper_units >= 47 or cargo["data"]["units"] >= cargo["data"]["capacity"]:
        print("‚úÖ  Cargo ready for delivery.")
        break

    # Cooldown wait
    cooldown = mine.get("data", {}).get("cooldown", {}).get("totalSeconds", 80)
    print(f"‚è≥  Cooldown: {cooldown}s\n")
    time.sleep(cooldown)
```
